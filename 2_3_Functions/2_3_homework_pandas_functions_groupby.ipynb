## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:
* оценка 2 и ниже — низкий рейтинг;
* оценка 4 и ниже — средний рейтинг;
* оценка 4.5 и 5 — высокий рейтинг.

Результат классификации запишите в столбец `class`.

```python
import pandas as pd

def classify_movies(movies, ratings):
    # Объединяем таблици movies и ratings
    joined = ratings.merge(movies, on='movieId', how='left')

    # Оставляем нужные столбцы
    filter_columns = joined[['movieId', 'rating', 'title']]

    # Вычисляем среднюю оценку для каждого фильма
    avg_ratings = filter_columns.groupby('movieId')['rating'].mean().reset_index()


    def classify_rating(avg_ratings):
        # Классифицируем оценки
        if avg_ratings <= 2:
            return 'низкий рейтинг'
        elif avg_ratings <= 4:
            return 'средний рейтинг'
        else:
            return 'высокий рейтинг'


    # Применяем классификацию оценок к столбцу rating
    avg_ratings['class'] = avg_ratings['rating'].apply(classify_rating)
    return avg_ratings

movies = pd.read_csv('/content/drive/MyDrive/PYDA/PYDA_HW9/ml-latest-small/movies.csv')
ratings = pd.read_csv('/content/drive/MyDrive/PYDA/PYDA_HW9/ml-latest-small/ratings.csv')

classified_movie_rating = classify_movies(movies, ratings)
classified_movie_rating
```

## Задание 2
Используйте файл `keywords.csv`.

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце `region` пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим `undefined`.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
```

Результат классификации запишите в отдельный столбец `region`.

```python
import pandas as pd


def geo_classifier(kw, geo_data):
    # Создаем столбец "region" со значением по умолчанию "undefined"
    kw['region'] = 'undefined'
    for k, v in geo_data.items():
      # Создаем маску для фильтрации по ключевым словам
      pattern = '|'.join(v)
      mask = kw['keyword'].str.contains(pattern, case=False)

      # Назначаем значение "region" для соответствующих строк
      kw.loc[mask, 'region'] = k

    return kw


kw = pd.read_csv('/content/drive/MyDrive/PYDA/PYDA_HW9/ml-latest-small/keywords.csv')
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

geo_classifier(kw, geo_data)
```
