#### Приведём размерности `fuel_consumption` в порядок.
⚠️ Колонки `fuel_consumption_l_100km` и `fuel_consumption_g_km` содержат множество значений в иных размерностях, нежели указано в описании к датасету. Для более точного анализа следует переместить каждую размерность в отдельную колонку. Однако эта задача занимает слишком много времени, поэтому принято решение удалить эти колонки из датафрейма.

В данных `fuel_consumption_l_100km`содержатся следующие примеры:
- `l/100 km` - оставить как есть
- `kg/100 km` - следует вынести в новую колонку `fuel_consumption_kg_100km`
- `kWh/100 km` - следует вынести в новую колонку `fuel_consumption_kWh_100km`
- `km (Ort)` - средний потребляемый объем топлива (обычно в литрах) на 100 километров пробега (дистанцию), основанного на стандартных условиях или типичных сценариях использования (Ort - от немецкого "ortsüblich", что можно перевести как "типичный" или "стандартный"). Можно считать равным `l/100 km`.

Аномалии:
- `0 l/100 km`
- `0 kWh/100 km`
- `- (l/100 km)`

Размерности в аномалиях соответствуют типу топлива. Вероятно при заполнении данных была допущена ошибка или сведения о расходе топлива не были получены. Рекомендуется изменить на `0`.

```python
anomaly = car_clear.copy()
anomaly = (anomaly.loc[(anomaly['fuel_consumption_l_100km']
                        .isin(['0 l/100 km']))])
t = anomaly.groupby(['fuel_type'])['fuel_consumption_l_100km'].agg(['count'])
t
```

```python
anomaly = car_clear.copy()
anomaly = (anomaly.loc[(anomaly['fuel_consumption_l_100km']
                        .isin(['0 kWh/100 km']))])
t = anomaly.groupby(['fuel_type'])['fuel_consumption_l_100km'].agg(['count'])
t
```

```python
anomaly = car_clear.copy()
anomaly = (anomaly.loc[(anomaly['fuel_consumption_l_100km']
                        .isin(['- (l/100 km)']))])
t = anomaly.groupby(['fuel_type'])['fuel_consumption_l_100km'].agg(['count'])
t
```

```python
# Перенесём значения kg/100 km в новую колонку
new_car_clear = car_clear.copy()

mask = new_car_clear['fuel_consumption_l_100km'].str.contains(' kg/100 km', na=False)
new_car_clear['fuel_consumption_kg_100km'] = new_car_clear[mask]['fuel_consumption_l_100km']
new_car_clear.loc[mask, 'fuel_consumption_l_100km'] = np.nan

# Перенесём значения kWh/100 km в новую колонку
mask1 = new_car_clear['fuel_consumption_l_100km'].str.contains(' kWh/100 km', na=False)
new_car_clear['fuel_consumption_kWh_100km'] = new_car_clear[mask1]['fuel_consumption_l_100km']
new_car_clear.loc[mask1, 'fuel_consumption_l_100km'] = np.nan
new_car_clear.info()
```

В данных `fuel_consumption_g_km` содержатся следующие примеры:

- `g/km` - не требует изменений
- `km Reichweite` - означает пробег, то есть максимальное расстояние, которое может пройти транспортное средство (например, электромобиль или гибридный автомобиль) на одной полной зарядке аккумулятора или единице топлива.

- `km (Ort)` - средний потребляемый объем топлива (обычно в литрах) на 100 километров пробега (дистанцию), основанного на стандартных условиях или типичных сценариях использования (Ort - от немецкого "ortsüblich", что можно перевести как "типичный" или "стандартный"). Можно считать равным `l/100 km`.
- `- (g/km)` - Вероятно при заполнении данных была допущена ошибка или сведения о расходе топлива не были получены. Рекомендуется изменить на `0`.

### 3. Как менялась популярность цветов у Mercedes-Benz с годами
Динамика популярности цветов автомобилей Mercedes-Benz показывает, что:
- Серебряный цвет вплоть до 2008 года был самым популярным цветом
- Резкий спад числа чёрных автомобилей в 2012 году может быть вызван тем, что в 2012 году компания выпустина меньше моделей чёрного цвета.

```python
mercedes = car_clear[car_clear['brand'] == 'mercedes-benz']
top_colors = ['black', 'silver', 'grey']
filtered_data = mercedes[mercedes['color'].isin(top_colors)]
color_maping = {'black':'y',
                'silver':'m',
                'grey':'c'}

# Группируем по году и цвету и считаем количество автомобилей
grouped = (filtered_data.groupby(['year', 'color'])
           .size().reset_index(name='count'))

fig, ax = plt.subplots(figsize=(12, 5))

# Строим график цветов для каждого года отдельно
for color, group in grouped.groupby('color'):
    plt.plot(group['year'], group['count'],
             label=str(color), color=color_maping.get(color))

ax.set_xlabel('Год производства')
ax.set_ylabel('Количество автомобилей')
ax.set_title('Динамика популярности цветов Mercedes-Benz')
ax.legend()
plt.tight_layout()
years = grouped['year'].unique()

# Устанавливаем явные значения и подписи для оси x
plt.xticks(years, map(str, years), rotation=45)
plt.show()
```

### 4. Как менялась популярность разных типов топлива
- Подавляющее большенство автомобилей используют традиционные виды топлива: бензин, дизель и гибридные установки (также на бензине или дизельном топливе)
- Доля автомобилей на дизельном топливе росла вплоть до 2016 года
- С 2016 года набирают популярность гибридные двигатели (совместное использование двигателя внутреннего сгорания и электродвигателя). Это может быть обусловлено развитием технологий электродвигателей и энергонакопителей
- Сжиженый газ (LPG) был популярен в качестве более экологичной альтернативы традиционным видам топлива, но в 2000-х годах стал уступать другим видам топлива, в т.ч. сжатому природному газу (CNG).
- С 2012 года резко вырастает доля электромобилей. Это может быть связано с быстрым темпом развития технологий электродвигателей и энергонакопителей, и развитием сервисной инфраструкруты для электромобилей. Также рост может быть связан с трендом на экологичность.
- В 2023 году вновь набирает популярность сжиженый газ. Это может быть следствием энергетического кризиса, усугубленного пандемией COVID-19 и сложной политической обстановкой, как ответная мера на увеличение цен на иные топливные нефтепродукты.

```python
car_clear['fuel_type'].unique()
```

Сформируем сводные таблицы для разных типов топлива
- `classic_fuel_type_per_year` - традиционные виды топлива (`Petrol`, `Diesel`, `Hybrid`, `Diesel Hybrid`)
- `eco_fuel_type_per_year` - экологичные виды топлива (`LPG`, `CNG`, `Electric`, `Ethanol`, `Hydrogen`)

```python
classic_fuel_type_per_year = (car_clear
      .groupby('year')['fuel_type']
      .value_counts(normalize=True)
      .mul(100)
      .round(2)
      .unstack())
classic_fuel_type_per_year = (classic_fuel_type_per_year
                              .drop(['LPG', 'CNG', 'Electric', 'Ethanol',
                                     'Hydrogen', 'Other', 'Unknown'], axis=1))

eco_fuel_type_per_year = (car_clear
      .groupby('year')['fuel_type']
      .value_counts(normalize=True)
      .mul(100)
      .round(2)
      .unstack())
eco_fuel_type_per_year = (eco_fuel_type_per_year
                          .drop(['Petrol', 'Diesel', 'Hybrid', 'Diesel Hybrid',
                                 'Other', 'Unknown'], axis=1))
```

```python
# Создаем stacked bar plot
classic_fuel_type_per_year.plot(kind='bar',
                                stacked=True,
                                figsize=(10, 6),
                                width=0.8,
                                colormap='Paired')

# Настройка осей и заголовка
plt.xlabel('Год выпуска')
plt.ylabel('Доля автомобилей, %')
plt.title('Доля автомобилей с традиционным типом топлива по годам')

# Отображаем легенду
plt.legend(loc='center left',
           bbox_to_anchor=(1, 0.5),
           labels=classic_fuel_type_per_year.columns,
           title='Типы топлива',
           facecolor='lightgray')

plt.show()
```

```python
# Создаем stacked bar plot
eco_fuel_type_per_year.plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8, colormap='Paired')

# Настройка осей и заголовка
plt.xlabel('Год выпуска')
plt.ylabel('Доля автомобилей, %')
plt.title('Доля автомобилей с экологичным типом топлива по годам')

# Отображаем легенду
plt.legend(loc='center left',
           bbox_to_anchor=(1, 0.5),
           labels=eco_fuel_type_per_year.columns,
           title='Типы топлива',
           facecolor='lightgray')

plt.show()
```
