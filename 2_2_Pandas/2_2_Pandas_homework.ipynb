## Задание 1
По данным файлов `movies.csv` и `ratings.csv` определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd


def rated_five_film(movies_file, rating_file):

    movies = pd.read_csv(movies_file)
    ratings = pd.read_csv(rating_file)

    # Оъединяем две таблицы csv по общему значению movieId
    joined = ratings.merge(movies, on='movieId', how='left')[['movieId', 'rating']]

    # Фильтруем полученную таблицу по оценке 5.0
    max_rate = joined.query(f'rating >= 5.0')

    # Находим id фильма, которому было выставлено больше всего оценок 5.0
    stat = max_rate['movieId'].value_counts().idxmax()

    return stat


movies_file = '/content/drive/MyDrive/PYDA/PYDA_HW_Pandas/HW_files_1/movies.csv'
rating_file = '/content/drive/MyDrive/PYDA/PYDA_HW_Pandas/HW_files_1/ratings.csv'

print(f'ID фильма с наибольшим числом оценок 5.0: {rated_five_film(movies_file, rating_file)}')
```

## Задание 2
По данным файла `power.csv` посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения `quantity`.

```python
import pandas as pd


def balt_consum(power_file):
    power = pd.read_csv(power_file)

    # Фильтруем даные по категориям, странам и годам.
    # Убираем отрицательные значения quantity
    filter = power[
        ((power.category == 4) | (power.category == 12) |
         (power.category == 21)) & ((power.country == 'Lithuania') |
         (power.country == 'Latvia') | (power.country == 'Estonia')) &
        ((power.year >= 2005) & (power.year <= 2010)) & (power.quantity > 0) ]

    return filter


power_file = '/content/drive/MyDrive/PYDA/PYDA_HW_Pandas/HW_files_2/power.csv'


print("Суммарное потребление стран Прибалтики: ")
print(balt_consum(power_file)['quantity'].sum())
```
